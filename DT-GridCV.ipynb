```python
dataset= pd.read_csv('loan_data.csv')
dataset
```

```python
dataset= pd.get_dummies(dataset,dtype=int, drop_first= True)
dataset
```

```python
indep= dataset[['person_age', 'person_income', 'person_emp_exp', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length',
       'credit_score', 'person_gender_male',
       'person_education_Bachelor', 'person_education_Doctorate',
       'person_education_High School', 'person_education_Master',
       'person_home_ownership_OTHER', 'person_home_ownership_OWN',
       'person_home_ownership_RENT', 'loan_intent_EDUCATION',
       'loan_intent_HOMEIMPROVEMENT', 'loan_intent_MEDICAL',
       'loan_intent_PERSONAL', 'loan_intent_VENTURE',
       'previous_loan_defaults_on_file_Yes']]
dep= dataset[['loan_status']]
```

```python
dataset['loan_status'].value_counts()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid= {'criterion':['gini', 'entropy', 'log_loss'],
              'max_features':['sqrt', 'log2', None],
             'splitter':['best', 'random']}

grid= GridSearchCV (DecisionTreeClassifier(),param_grid,refit = True, verbose = 0,n_jobs=None,scoring='f1_weighted')

grid.fit(indep,dep)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(indep) 
```

```python
from sklearn.metrics import confusion_matrix
confusion= confusion_matrix(dep,grid_predictions)

```

```python
print(confusion)
```

```python
from sklearn.metrics import classification_report
class_report= classification_report(dep,grid_predictions)
```

```python
print(class_report)
```
